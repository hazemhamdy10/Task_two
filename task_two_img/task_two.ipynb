```python
import numpy as np 
import matplotlib.pyplot as plt 
import cv2 
```

1- using sobel for edge detection

```python
img=cv2.imread("lenna.jpg")
original_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.subplot(221)
plt.imshow(original_img,cmap='gray')
plt.title("original")
plt.axis('off')

gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.subplot(222)
plt.imshow(gray,cmap='gray')
plt.title("Gray")
plt.axis('off')

blur=cv2.GaussianBlur(gray,(15,15),100)
plt.subplot(223)
plt.imshow(blur,cmap='gray')
plt.title("blured")
plt.axis('off')

sobel_X=cv2.Sobel(blur,-1,1,0)
sobel_Y=cv2.Sobel(blur,-1,0,1)

sobeled_img=cv2.add(sobel_X,sobel_Y)
plt.subplot(224)
plt.imshow(sobeled_img,cmap='gray')
plt.title("sobel_edge_detection")
plt.axis('off')


plt.show()



```

2-using canny for edge detection (best one )

```python
img=cv2.imread("lenna.jpg")
original_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.subplot(221)
plt.imshow(original_img,cmap='gray')
plt.title("original")
plt.axis('off')

gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.subplot(222)
plt.imshow(gray,cmap='gray')
plt.title("Gray")
plt.axis('off')

blur=cv2.GaussianBlur(gray,(15,15),100)
plt.subplot(223)
plt.imshow(blur,cmap='gray')
plt.title("blured")
plt.axis('off')

canny_image=cv2.Canny(blur,22,50)
plt.subplot(224)
plt.imshow(canny_image,cmap='gray')
plt.title("canny_edge_detection")
plt.axis('off')
```

3-Laplacian edge detection 

```python
img=cv2.imread("lenna.jpg")
original_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.subplot(221)
plt.imshow(original_img,cmap='gray')
plt.title("original")
plt.axis('off')

gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.subplot(222)
plt.imshow(gray,cmap='gray')
plt.title("Gray")
plt.axis('off')

blur=cv2.GaussianBlur(gray,(15,15),100)
plt.subplot(223)
plt.imshow(blur,cmap='gray')
plt.title("blured")
plt.axis('off')

laplacian_image = cv2.Laplacian(blur, -1)
plt.subplot(224)
plt.imshow(laplacian_image,cmap='gray')
plt.title("laplacian_edge_detection")
plt.axis('off')
```
